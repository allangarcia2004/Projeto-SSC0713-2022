```python
import subprocess
import json
import os
import shlex
```

```python
PYTHON = './venv/bin/python'
SCRIPT = './src/main.py'
LOGS_FOLDER = './logs/'
```

```python
neurons_disposition = [2, 3, 2]
population_size = 250
tournament_size = population_size // 10
crossover_probability = 0.3
mutation_probability = 0.4
fitness_weights = [10, -1]
```

```python
def run_case(test_name: str,
             neurons_disposition=neurons_disposition,
             population_size=population_size,
             crossover_probability=crossover_probability,
             mutation_probability=mutation_probability,
             fitness_weights=fitness_weights,
             tournament_size=tournament_size):
    logbook_file = os.path.join(LOGS_FOLDER, 'pkl', test_name + '.pkl')
    args_file = os.path.join(LOGS_FOLDER, 'args', test_name + '.json')

    os.makedirs(os.path.dirname(logbook_file), exist_ok=True)
    os.makedirs(os.path.dirname(args_file), exist_ok=True)

    arguments = {
        'use_backup': False,
        'neurons_disposition': neurons_disposition,
        'population_size': population_size,
        'hall_of_fame_size': 1,
        'tournament_size': tournament_size,
        'crossover_probability': crossover_probability,
        'mutation_probability': mutation_probability,
        'generations': 200,
        'fitness_weights': fitness_weights,
        'logbook_file': logbook_file
    }

    with open(args_file, 'w') as file:
        json.dump(arguments, file, indent=4)

    command = shlex.join([PYTHON, SCRIPT, json.dumps(arguments)])
    return subprocess.Popen(command, shell=True)
```

```python
cases = []
```

```python
population_sizes = [10, 50, 100, 200, 500, 1000]

for x in population_sizes:
    test_name = f'population_size/{x}'
    case = {'population_size': x, 'test_name': test_name}
    cases.append(case)
```

```python
dispositions = [[2, 2], [2, 4, 2], [2, 3, 3, 2]]

for x in dispositions:
    test_name = f'neurons_disposition/{x}'
    case = {'neurons_disposition': x, 'test_name': test_name}
    cases.append(case)
```

```python
tournament_size_divisions = [2, 10, 20]

for x in tournament_size_divisions:
    test_name = f'tournament_size/{x}'
    case = {'tournament_size': tournament_size // x, 'test_name': test_name}
    cases.append(case)
```

```python
probabilities = [.25, .5, .75]

for x in probabilities:
    test_name = f'crossover_probability/{str(x).replace(".",",")}'
    case = {'crossover_probability': x, 'test_name': test_name}
    cases.append(case)
```

```python
probabilities = [.25, .5, .75]

for x in probabilities:
    test_name = f'mutation_probability/{str(x).replace(".",",")}'
    case = {'mutation_probability': x, 'test_name': test_name}
    cases.append(case)
```

```python
weights = [(1e3, -1e-7), (2, -1), (10, -1), (1e2, -1), (1e3, -1)]

for x in weights:
    a, b = x
    a, b = float(a), float(b)
    x = a, b

    test_name = f'fitness_weights/{x}'
    case = {'fitness_weights': x, 'test_name': test_name}
    cases.append(case)
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
for chunk in chunks(cases, 4):
    processes = [run_case(**case) for case in chunk]
    for p in processes:
        p.wait()
```
