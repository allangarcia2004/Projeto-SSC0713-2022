```python
import pickle

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set()
```

```python
with open("statistics_by_generation.pickle", "rb") as file:
    data = pickle.load(file)
```

```python
df = pd.DataFrame.from_records(data)
```

```python
labels = []
x_axis = range(len(df))

fig, ax = plt.subplots()

for col in df.columns:
    if col == "std":
        continue
    if col == "average":
        ax.errorbar(x_axis, df[col], df["std"])
    else:
        ax.plot(x_axis, df[col])
    labels.append(col)

ax.set_xlim(1, len(df))
plt.show()
```
